```python
filtered_indicators = []
first_word = ''
for indicator, data in grp_indicators:
    if first_word == indicator.split()[0]:
        # filtered_indicators.append(indicator)
        continue
    else:
        first_word = indicator.split()[0]
        filtered_indicators.append(indicator)
        
print(len(filtered_indicators))
for i in filtered_indicators: 
    print(i)
```

```python
indicators_filtered = ['GDP growth (annual %)',
                       'Life expectancy at birth, total (years)',
                       'Number of infant deaths',
                       'Age dependency ratio (% of working-age population)',
                       'Literacy rate, adult total (% of people ages 15 and above)',
                      #  'Income share held by fourth 20%',
                      #  'Poverty gap at $2.15 a day (2017 PPP) (%)',
                      #  'Human capital index (HCI) (scale 0-1)',
                      #  'Research and development expenditure (% of GDP)',
                       'Electric power consumption (kWh per capita)',
                       'Renewable electricity output (% of total electricity output)',
                       'Forest area (% of land area)',
                       'Arable land (% of land area)',
                       'Carbon dioxide (CO2) emissions (total) excluding LULUCF (% change from 1990)',
                       'Exports of goods and services (% of GDP)'
                       ]

countries_filtered = [
                        "United States",
                        "United Kingdom",
                        "Germany",
                        "China",
                        "Japan",
                        "India",
                        "Brazil",
                        "Indonesia",
                        "South Africa",
                        "Malaysia",
                        "Niger",
                        "Chad",
                        "Haiti",
                        # "South Sudan",
                        "Burundi"
                      ]

print(len(indicators_filtered))
print(len(countries_filtered))
```

```python
data_wdi_filtered = data_wdi[data_wdi['Indicator Name'].isin(indicators_filtered) & data_wdi['Country Name'].isin(countries_filtered)]
print(data_wdi_filtered.shape)
print(data_wdi_filtered)
data_wdi_filtered.to_csv('data_wdi_filtered.csv', index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme(style="whitegrid")
plot_data = pd.read_csv('data_wdi_filtered.csv')
plot_data_filtered = plot_data.groupby('Indicator Name')


for indicator, data1 in plot_data_filtered:
    print("Indicator:", indicator)
    plt.figure(figsize=(12, 8))
    
    for idx, (country, data2) in enumerate(data1.groupby('Country Name')):
        data_plot = data2.iloc[:, -24:-1].to_numpy().squeeze()
        sns.lineplot(data=data_plot, label=country)
    
    plt.title(indicator, fontsize=16, weight='bold')
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('Value', fontsize=14)
    plt.legend(title='Country', fontsize=12, title_fontsize=12, loc='best')
    plt.xticks(rotation=45, fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout() 

plt.show()
```
