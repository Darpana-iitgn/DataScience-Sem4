```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
import numpy as np
import sys
import time

# --- Configuration ---
DATA_FILE_PATH = 'poverty_bpl_ptage.csv' # <-- IMPORTANT: Change this to your POVERTY data CSV file path
GEO_NAME_COL = geo_name           # Column containing the geographical entity name
AVAIL_YEARS_COL = avail_yrs       # Column containing the JSON string of available years
VALUE_PREFIX = 'Value '             # Prefix for the value columns (e.g., 'Value 2000')

MIN_YEAR = 1950
MAX_YEAR_TARGET = 2050 # The end year for the final filled series
TOTAL_YEARS = MAX_YEAR_TARGET - MIN_YEAR + 1

# Polynomial Fitting Parameters
POLYNOMIAL_DEGREE = 10
# Minimum number of data points required to attempt polynomial fitting
# A degree 'd' polynomial has d+1 coefficients. Need at least d+1 points.
# Using d+2 or more is generally safer to avoid perfect fitting to noise.
MIN_POINTS_FOR_POLYFIT = POLYNOMIAL_DEGREE + 1 # Require at least 7 points for deg 5

# Plotting Parameters
PLOT_FIG_SIZE = (15, 10)
PLOT_TITLE = f"Poverty Trend ({MIN_YEAR}-{MAX_YEAR_TARGET}) with Polynomial Fit (Degree {POLYNOMIAL_DEGREE})"
PLOT_X_LABEL = "Year"
PLOT_Y_LABEL = "Poverty Value"

# Reproducibility
SEED = 42
np.random.seed(SEED)
# --- End Configuration ---


# --- Helper Function ---
def interpolate_with_polynomial(years, values, geo_name):
    """
    Fills missing values in the 'values' array using polynomial fitting or linear interpolation.
    Returns the filled array.
    """
    filled_values = values.copy() # Work on a copy
    all_years_np = np.arange(MIN_YEAR, MAX_YEAR_TARGET + 1)

    # Find indices and corresponding years/values where data is known (not NaN)
    known_indices = np.where(~np.isnan(values))[0]
    known_years = all_years_np[known_indices]
    known_values = values[known_indices]

    num_known_points = len(known_indices)

    if num_known_points == 0:
        print(f"  Warning: No data points found for {geo_name}. Cannot interpolate.")
        return filled_values # Return array of NaNs
    elif num_known_points == 1:
        print(f"  Warning: Only one data point for {geo_name}. Filling all years with this value.")
        filled_values[:] = known_values[0]
        return filled_values
    elif num_known_points < MIN_POINTS_FOR_POLYFIT:
        # Not enough points for the desired polynomial degree, fall back to linear interpolation
        print(f"  Info: Using linear interpolation for {geo_name} (found {num_known_points} points, need {MIN_POINTS_FOR_POLYFIT} for poly fit).")
        # Use numpy's interpolation. It handles cases with 2+ points.
        # Note: np.interp does not extrapolate beyond the range of known_indices by default.
        # To fill the whole range, we interpolate over all_indices based on known_indices/known_values.
        all_indices = np.arange(TOTAL_YEARS)
        interpolated_linear = np.interp(all_indices, known_indices, known_values)
        # Fill only the originally NaN values with the linear interpolation
        nan_indices = np.where(np.isnan(filled_values))[0]
        filled_values[nan_indices] = interpolated_linear[nan_indices]

    else:
        # Sufficient points for polynomial fitting
        print(f"  Info: Using Polynomial Fitting (degree {POLYNOMIAL_DEGREE}) for {geo_name} (found {num_known_points} points).")
        try:
            # Fit the polynomial to the known data points
            # polyfit returns coefficients [c_n, c_{n-1}, ..., c_0] for c_n*x^n + ... + c_0
            coeffs = np.polyfit(known_years, known_values, POLYNOMIAL_DEGREE)

            # Create a polynomial function from the coefficients
            poly_func = np.poly1d(coeffs)

            # Predict values for all years in the target range
            predictions = poly_func(all_years_np)

            # Fill missing values (NaNs) with the predictions
            # It's generally better to only fill the NaNs rather than overwriting known points
            nan_indices = np.where(np.isnan(filled_values))[0]
            filled_values[nan_indices] = predictions[nan_indices]

            # Optional: Check for large extrapolation values and potentially cap them
            # if np.any(np.abs(predictions[nan_indices]) > some_threshold):
            #    print(f"  Warning: Large extrapolation values predicted for {geo_name}. Consider capping or reviewing fit.")
            #    # Example capping:
            #    # filled_values = np.clip(filled_values, known_values.min()*0.8, known_values.max()*1.2)

        except (np.linalg.LinAlgError, ValueError) as e:
             print(f"  Error: Polynomial fitting failed for {geo_name}. Error: {e}. Falling back to linear interpolation.")
             # Fallback logic copied from the linear interpolation block above
             all_indices = np.arange(TOTAL_YEARS)
             interpolated_linear = np.interp(all_indices, known_indices, known_values)
             nan_indices = np.where(np.isnan(filled_values))[0]
             filled_values[nan_indices] = interpolated_linear[nan_indices]


    return filled_values

# --- Main Processing Logic ---
processed_data = {} # Dictionary to store results: {geo_name: filled_values_array}

# Load the dataset
try:
    df_poverty = pd.read_csv(DATA_FILE_PATH)
    print(f"Successfully loaded data from '{DATA_FILE_PATH}'. Shape: {df_poverty.shape}")
    # Verify required columns exist
    required_cols = [GEO_NAME_COL, AVAIL_YEARS_COL]
    if not all(col in df_poverty.columns for col in required_cols):
         print(f"Error: Missing required columns. Need '{GEO_NAME_COL}' and '{AVAIL_YEARS_COL}'.")
         print(f"Available columns: {list(df_poverty.columns)}")
         sys.exit(1)

except FileNotFoundError:
    print(f"Error: Data file not found at '{DATA_FILE_PATH}'")
    sys.exit(1)
except Exception as e:
    print(f"Error loading data: {e}")
    sys.exit(1)


# Iterate over each geographical entity
print(f"\nProcessing {len(df_poverty)} geographical entities...")
start_process_time = time.time()

for idx, row in df_poverty.iterrows():
    geo_name = row[GEO_NAME_COL]
    # print(f"\nProcessing: {geo_name} ({idx+1}/{len(df_poverty)})") # Less verbose output

    # Initialize array for the full time range
    values_over_time = np.full(TOTAL_YEARS, np.nan)

    # Parse available years safely
    try:
        available_years = json.loads(row[AVAIL_YEARS_COL])
        if not isinstance(available_years, list):
             # print(f"  Warning: Parsed 'avail_yrs' for {geo_name} is not a list. Skipping.") # Less verbose
             continue
    except (json.JSONDecodeError, TypeError):
        # print(f"  Warning: Could not parse 'avail_yrs' JSON for {geo_name}. Skipping.") # Less verbose
        continue

    # Populate known values
    valid_data_points = 0
    min_available_year_in_range = MAX_YEAR_TARGET + 1
    max_available_year_in_range = MIN_YEAR - 1
    for year in available_years:
        if MIN_YEAR <= year <= MAX_YEAR_TARGET:
            value_col_name = f"{VALUE_PREFIX}{year}"
            if value_col_name in row and pd.notna(row[value_col_name]): # Check for NaN using pandas
                try:
                    value = float(row[value_col_name])
                    values_over_time[year - MIN_YEAR] = value
                    valid_data_points += 1
                    min_available_year_in_range = min(min_available_year_in_range, year)
                    max_available_year_in_range = max(max_available_year_in_range, year)
                except (ValueError, TypeError):
                     pass # Ignore conversion errors silently for now
                     # print(f"  Warning: Could not convert value '{row[value_col_name]}' for year {year} in {geo_name} to float.")


    if valid_data_points == 0:
        # print(f"  Skipping {geo_name} due to no valid data points within the specified range and columns.") # Less verbose
        continue

    # Interpolate/Extrapolate using the helper function
    filled_values = interpolate_with_polynomial(np.arange(MIN_YEAR, MAX_YEAR_TARGET + 1), values_over_time, geo_name)

    # Store the results
    processed_data[geo_name] = filled_values

    # Optional: Print progress periodically
    if (idx + 1) % 50 == 0 or (idx + 1) == len(df_poverty):
         print(f"Processed {idx+1}/{len(df_poverty)} entities...")


print(f"\nData processing completed in {time.time() - start_process_time:.2f} seconds.")

# --- Plotting ---
if not processed_data:
    print("\nNo data was processed successfully. Cannot generate plot.")
else:
    print("\nGenerating plot...")
    plt.figure(figsize=PLOT_FIG_SIZE)
    sns.set_style("whitegrid")

    plot_years = np.arange(MIN_YEAR, MAX_YEAR_TARGET + 1)

    for geo_name, values in processed_data.items():
        if not np.all(np.isnan(values)):
             sns.lineplot(x=plot_years, y=values, label=geo_name, linewidth=1.0, alpha=0.8) # Adjust line appearance
        # else: # Don't print skip message for every empty plot
             # print(f"Skipping plot for {geo_name} as all values are NaN after processing.")

    plt.xlabel(PLOT_X_LABEL)
    plt.ylabel(PLOT_Y_LABEL)
    plt.title(PLOT_TITLE)
    plt.ylim(bottom=0) # Often poverty metrics don't go below zero

    # Improve legend handling - potentially show fewer labels if too many
    handles, labels = plt.gca().get_legend_handles_labels()
    num_labels = len(labels)
    if num_labels > 30: # Example threshold: if more than 30 lines, don't show legend
        print(f"Hiding legend because there are {num_labels} items.")
        plt.legend().set_visible(False)
    elif num_labels > 0 :
        plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0., fontsize='small')
        plt.tight_layout(rect=[0, 0, 0.85, 1]) # Adjust layout further if legend is wide
    else:
        plt.tight_layout()


    plt.show()
    print("Plot generation complete.")
```
